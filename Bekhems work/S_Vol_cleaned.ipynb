```python
import pandas as pd
import numpy as np
#import warnings
#warnings.filterwarnings("ignore")
```

```python
# The path to CSV file
file = 'The_Volcanoes_Of_Earth.csv'
```

```python
# Read volcano data into pandas
df = pd.read_csv(file)
df.head()
```

```python
# Get a list of all of columns for easy reference
df.columns
```

```python
df.info()
```

```python
df.Latitude=df.Latitude.apply(lambda x: "+" + str(x) if "N" in str(x) else "-" + str(x))
df.Longitude=df.Longitude.apply(lambda x: "+" + str(x) if "E" in str(x) else "-" + str(x))
df.head()
```

```python
# Remove degree symbol from Latitude and Longitude
df['Latitude'] = df.Latitude.astype(str).str.replace('°N','')
df['Latitude'] = df.Latitude.astype(str).str.replace('°S' , '')
df['Longitude'] = df.Longitude.astype(str).str.replace('°E' , '')
df['Longitude'] = df.Longitude.astype(str).str.replace('°W' , '')
df.head()
```

```python
# Change Latitude and Longidue to a float 
df['Latitude'] = df.Latitude.astype(float)
df['Longitude'] = df.Longitude.astype(float)
```

```python
df.head()
```

```python
# Split summit and elevation
df[['Summit', 'Elevatiuon']] = df['Summit_and_Elevatiuon'].str.split(',', expand=True)
```

```python
df['Summit'] = df.Summit.astype(str).str.replace(' m' , '')
df['Summit'] = df.Summit.astype(str).str.strip()
df['Elevatiuon'] = df.Elevatiuon.astype(str).str.replace(' ft' , '')
df['Elevatiuon'] = df.Elevatiuon.astype(str).str.strip()
```

```python
# Rename column due to misspelling
df.rename(columns = {'Elevatiuon':'Elevation'}, inplace = True)
```

```python
# Drop null values
cols = ['population_within_5km','population_within_10km','population_within_30km','population_within_100km']
df[cols] = df[cols].replace({'0':np.nan, 0:np.nan})
df2=df.dropna(subset=['population_within_5km','population_within_10km','population_within_30km','population_within_100km'])
```

```python
# Drop column
df2.drop('Summit_and_Elevatiuon', axis=1,inplace=True)
```

```python
# Drop column
df2.drop('Volcano_Image', axis=1,inplace=True)
```

```python
df2.head()
```

```python
df2.Volcano_Type.value_counts()
```

```python
df2["Volcano_Type"] = df2.Volcano_Type.replace("Stratovolcano(es)", "Stratovolcano")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Pyroclastic cone(s)", "Pyroclastic cone")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Lava dome(s)", "Lava dome")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Shield(s)", "Shield")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Fissure vent(s)", "Fissure vent")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Maar(s)", "Maar")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Tuff cone(s)", "Tuff cone")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Lava cone(es)", "Lava cone")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Lava cone(s)", "Lava cone")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Explosion crater(s)", "Explosion crater")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Stratovolcano?", "Stratovolcano")
df2["Volcano_Type"] = df2.Volcano_Type.replace("Caldera(s)", "Caldera")

df2.Volcano_Type.value_counts()
```
